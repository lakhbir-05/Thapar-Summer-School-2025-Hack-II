```python
import pandas as pd
import numpy as np
from pycaret.regression import *
from sklearn.ensemble import GradientBoostingRegressor

# Load data
train = pd.read_csv("/content/train.csv")
test = pd.read_csv("/content/test (1).csv")

# Drop unnecessary columns
train = train.drop(columns=['Row#'])
test = test.drop(columns=['Row#'])

# Manual correlation-based feature selection
correlation_matrix = train.corr()
correlation_with_target = correlation_matrix['yield'].abs().sort_values(ascending=False)

# Select features with correlation > 0.1 (you can change threshold)
selected_features = correlation_with_target[correlation_with_target > 0.1].index.tolist()
selected_features.remove('yield')  # remove target from feature list

# Filter data
train_filtered = train[selected_features + ['yield']]
test_filtered = test[selected_features]

# PyCaret setup
setup(data=train_filtered,
      target='yield',
      remove_outliers=True,
      outliers_threshold=0.05,
      train_size=0.9,
      fold=3,
      session_id=42,
      verbose=False)

# Define model
gbr = GradientBoostingRegressor(random_state=42)

# Tune model
tuned_model = tune_model(
    gbr,
    optimize='MAE',
    search_library='optuna',
    search_algorithm='random',
    n_iter=50,
    choose_better=True,
    custom_grid={
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 4, 5],
        'min_samples_split': [2, 5, 10]
    }
)

# Finalize model
final_model = finalize_model(tuned_model)

# Predict
test_ids = test['id']
predictions = predict_model(final_model, data=test_filtered)

# Save submission
submission = pd.DataFrame({
    'id': test_ids,
    'yield': predictions['prediction_label']
})
submission.to_csv("Lakhbir_Kaur_1125.csv", index=False)
print("✅ Lakhbir_Kaur_1125.csv created with manual correlation-based feature selection!")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error

# Load datasets
train_df = pd.read_csv("/content/train.csv")
test_df = pd.read_csv("/content/test (1).csv")
sample_submission = pd.read_csv("/content/sample_submission.csv")

# Step 1: Drop 'id' and 'Row#' ONLY from the train data
train_df = train_df.drop(['id', 'Row#'], axis=1)

# Step 2: Remove outliers using quantile method
Q1 = train_df.quantile(0.01)
Q3 = train_df.quantile(0.99)
train_df = train_df[~((train_df < Q1) | (train_df > Q3)).any(axis=1)]

# Step 3: Select top 10 features correlated with 'yield'
correlations = train_df.corr()['yield'].abs().sort_values(ascending=False)
top_features = correlations.index[1:11]  # Exclude 'yield' itself
print("Top 10 correlated features:\n", top_features)

# Step 4: Prepare data
X = train_df[top_features]
y = train_df['yield']
X_test = test_df[top_features]  # Use the same features in test

# Step 5: Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(X_test)

# Step 6: Apply PCA
pca = PCA(n_components=0.95)  # Retain 95% variance
X_pca = pca.fit_transform(X_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Step 7: Train/test split and model training
X_train, X_valid, y_train, y_valid = train_test_split(X_pca, y, test_size=0.1, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate on validation set
y_pred = model.predict(X_valid)
print("Validation MSE:", mean_squared_error(y_valid, y_pred))

# Step 8: Retrain on full data and predict
model.fit(X_pca, y)
test_preds = model.predict(X_test_pca)

# Step 9: Save submission
sample_submission['yield'] = test_preds
sample_submission.to_csv('lakhbir_8788.csv', index=False)
print("Submission file saved as: lakhbir_8788.csv")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error

# Load datasets
train_df = pd.read_csv("/content/train.csv")
test_df = pd.read_csv("/content/test (1).csv")
sample_submission = pd.read_csv("/content/sample_submission.csv")

# Drop unused columns
X = train_df.drop(['id', 'Row#', 'yield'], axis=1)
y = train_df['yield']
X_test = test_df.drop(['id', 'Row#'], axis=1)

# Optional: Split for validation (comment this block when training on full data)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.1, random_state=42)

# Define the model
gbr = GradientBoostingRegressor(random_state=42)

# Define hyperparameter grid
param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 4],
    'min_samples_split': [2, 5],
}

# Perform Grid Search with cross-validation
grid_search = GridSearchCV(gbr, param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Get best model
best_model = grid_search.best_estimator_

# Evaluate on validation set
valid_preds = best_model.predict(X_valid)
print("Validation MSE:", mean_squared_error(y_valid, valid_preds))

# Retrain on full training data
best_model.fit(X, y)

# Predict on test data
test_preds = best_model.predict(X_test)

# --- Post-processing: Remove outliers using IQR ---
Q1 = np.percentile(test_preds, 25)
Q3 = np.percentile(test_preds, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Clip outliers
test_preds_processed = np.clip(test_preds, lower_bound, upper_bound)

# Save final predictions
sample_submission['yield'] = test_preds_processed
sample_submission.to_csv('lakhbir_9944.csv', index=False)
print("Submission saved as lakhbir_9944.csv")
```

```python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import optuna

# Load datasets
train_df = pd.read_csv("/content/train.csv")
test_df = pd.read_csv("/content/test (1).csv")
sample_submission = pd.read_csv("/content/sample_submission.csv")

# Prepare features and target
X = train_df.drop(['id', 'Row#', 'yield'], axis=1)
y = train_df['yield']
X_test = test_df.drop(['id', 'Row#'], axis=1)

# Split for validation
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.1, random_state=42)

# Optuna objective function using MAE
def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 300),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'max_depth': trial.suggest_int('max_depth', 3, 6),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 10),
    }
    model = GradientBoostingRegressor(random_state=42, **params)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    mae = mean_absolute_error(y_valid, preds)
    return mae  # Optuna will minimize this

# Run Optuna optimization
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=200)  # Increase trials for better results

# Output best params and score
print("Best parameters:", study.best_params)
print("Best MAE:", study.best_value)

# Train best model on full data
best_model = GradientBoostingRegressor(random_state=42, **study.best_params)
best_model.fit(X, y)

# Predict on test set
test_preds = best_model.predict(X_test)

# Save submission
sample_submission['yield'] = test_preds
sample_submission.to_csv('lakhbir_7976.csv', index=False)
print("Saved: lakhbir_7976.csv")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import optuna

# Load datasets
train_df = pd.read_csv("/content/train.csv")
test_df = pd.read_csv("/content/test (1).csv")
sample_submission = pd.read_csv("/content/sample_submission.csv")

# Prepare data
X = train_df.drop(['id', 'Row#', 'yield'], axis=1)
y = train_df['yield']
X_test = test_df.drop(['id', 'Row#'], axis=1)

# Optional: Split for validation
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.1, random_state=42)

# Define Optuna objective
def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 300),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'max_depth': trial.suggest_int('max_depth', 3, 6),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 10),
    }
    model = GradientBoostingRegressor(random_state=42, **params)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    mae = mean_absolute_error(y_valid, preds)
    return mae  # minimizing MAE

# Run Optuna optimization
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=120)  # Adjust trials as needed

# Print best hyperparameters
print("Best hyperparameters:", study.best_params)

# Retrain best model on full training data
best_model = GradientBoostingRegressor(random_state=42, **study.best_params)
best_model.fit(X, y)

# Predict on test data
test_preds = best_model.predict(X_test)

# --- Post-processing: Remove outliers using IQR ---
Q1 = np.percentile(test_preds, 25)
Q3 = np.percentile(test_preds, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

test_preds_processed = np.clip(test_preds, lower_bound, upper_bound)

# Save final predictions
sample_submission['yield'] = test_preds_processed
sample_submission.to_csv('lakhbir_2944.csv', index=False)
print("Submission saved as lakhbir_2944.csv")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.metrics import mean_absolute_error

# Load datasets
train_df = pd.read_csv("/content/train.csv")
test_df = pd.read_csv("/content/test (1).csv")
sample_submission = pd.read_csv("/content/sample_submission (1).csv")

# Prepare features and target
X = train_df.drop(['id', 'Row#', 'yield'], axis=1)
y = train_df['yield']
X_test = test_df.drop(['id', 'Row#'], axis=1)

# Define the model
gbr = GradientBoostingRegressor(random_state=42)

# Hyperparameter grid
param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 4],
    'min_samples_split': [2, 5],
}

# Define KFold strategy
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Grid search with MAE as the scoring metric
grid_search = GridSearchCV(estimator=gbr,
                           param_grid=param_grid,
                           cv=kf,
                           scoring='neg_mean_absolute_error',
                           n_jobs=-1)
grid_search.fit(X, y)

# Best model and parameters
best_model = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)

# Calculate CV MAE scores (negated back to positive)
cv_mae_scores = -grid_search.cv_results_['mean_test_score']
print("Average CV MAE:", np.mean(cv_mae_scores))

# Train final model on all data
best_model.fit(X, y)

# Predict on test set
test_preds = best_model.predict(X_test)

# Save predictions
sample_submission['yield'] = test_preds
sample_submission.to_csv('lakhbir_6468.csv', index=False)
print("lakhbir_6468.csv created!")
```

```python
import pandas as pd
import numpy as np
from lightgbm import LGBMRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error

# Load datasets
train_df = pd.read_csv("/content/train.csv")
test_df = pd.read_csv("/content/test (1).csv")
sample_submission = pd.read_csv("/content/sample_submission (1).csv")

# Prepare features and target
X = train_df.drop(['id', 'Row#', 'yield'], axis=1)
y = train_df['yield']
X_test = test_df.drop(['id', 'Row#'], axis=1)

# KFold CV setup
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Store MAEs and predictions for averaging
mae_per_fold = []
test_preds = np.zeros(len(X_test))

# K-Fold loop
for fold, (train_idx, valid_idx) in enumerate(kf.split(X)):
    print(f"\nFold {fold + 1}")

    # Split data
    X_train, X_valid = X.iloc[train_idx], X.iloc[valid_idx]
    y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]

    # Define LightGBM model with fixed parameters
    model = LGBMRegressor(
        n_estimators=600,
        learning_rate=0.03,
        max_depth=9,
        objective='mae',
        random_state=fold
    )

    # Train and predict
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    mae = mean_absolute_error(y_valid, y_pred)
    mae_per_fold.append(mae)
    print(f"MAE: {mae:.4f}")

    # Predict test set and accumulate predictions
    test_preds += model.predict(X_test) / kf.n_splits

# Print all fold MAEs
print("\nMAE per fold:")
for i, mae in enumerate(mae_per_fold, 1):
    print(f"Fold {i}: MAE = {mae:.4f}")
print(f"\nAverage MAE: {np.mean(mae_per_fold):.4f}")

# Save final predictions
sample_submission['yield'] = test_preds
sample_submission.to_csv('lakhbir_1468.csv', index=False)
print("\nlakhbir_1468.csv created!")
```
